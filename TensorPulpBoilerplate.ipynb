# Imports for Python and R

### Step 1: Import Python Dependencies.
GrapefruitDataFrame is generated in the process for testing. You may need to pip install if the packages are not on your machine yet.

```python
# Select which packages to install
# pip install pandas
# pip install rpy2

import pandas as pd
import rpy2 as rp

GrapefruitDataFrame = pd.DataFrame({ 'ID': [1,2,3,4,5],
                                    'IV':[2,4,5,8,10],
                                    'DV':[5, 10, 15, 21, 25]})
```

### Step 2: Load in the `rpy2.ipython` Extension
This will allow us to interface with R in Python.

```python
%load_ext rpy2.ipython
```

### Step 3: In R, Import (`-i`) the Test DataFrame and Install `lavaan`
Contains additional code to set the CRAN cloud mirror as the default package repo.

```python
%%R -i GrapefruitDataFrame
options(repos = c(CRAN = "https://cloud.r-project.org")) # Set global CRAN cloud mirror as default package repo
install.packages("lavaan")
library(lavaan)
```

# Begin Structural Equation Models
Sample code is provided.

```python
%%R -i GrapefruitDataFrame
ModelCode <- 'DV ~ ID + IV'
fit <- sem(model = ModelCode, data = GrapefruitDataFrame)
summary(fit, standardized = TRUE, fit.measures= TRUE)
```
